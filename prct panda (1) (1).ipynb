```python
# Region wise total Sales

# Region
# Sales
# Operation - sum (Sales)
```

```python
df.groupby('Region').Sales.sum()
```

```python
pd.pivot_table(df, values='Sales',index='Region', aggfunc='sum')
```

```python
# Region wise total Sales of each product

# Region
# Sales
# Type
# Operation - sum (Sales)
```

```python
pd.pivot_table(df, values='Sales',index='Type',columns='Region', aggfunc='sum',fill_value=0)
```

```python
pd.pivot_table(df, values='Sales',index=['Type','Region'],aggfunc='sum',fill_value=0)
```

```python
# Region wise total Sales of each product sold by each Salesperson

# Region
# Sales
# Type
# Salesperson
# Operation - sum (Sales)
```

```python
pd.pivot_table(df, values='Sales',index=['Type','Region'],columns='Salesperson',aggfunc='sum')
```
