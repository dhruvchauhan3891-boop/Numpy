# Indexing and Slicing of a Dataframe

### Indexing

```python
### Implicit Indexes
# df.iloc[row_index]
# df.iloc[row_index, col_index]

### Explicit Indexes
# df.loc[row_index]
# df.loc[row_index, col_index]
```

```python
df
```

```python
# Fisrt row.
df.iloc[0]
```

```python
# Fisrt row.
df.iloc[[0]]
```

```python
# Last Row.
df.iloc[-1]
```

```python
# Last Row.
df.iloc[[-1]]
```

```python
df.index=list('abcdefgh')
```

```python
df
```

```python
# Get the details of index b

df.loc[['b']]
```

```python
# Get the city at index g

df.loc['g', 'City']
```

```python
# First row last column.

df.iloc[0, -1]
```

### Slicing

```python
# ### Implicit Indexes
# df.iloc[row_slicing]
# df.iloc[start:stop:step]

# df.iloc[row_slicing, col_slicing]
# df.iloc[start:stop:step, start:stop:step]

# (stop_index is excluded)

# ### Explicit Indexes
# df.loc[row_slicing]
# df.loc[start:stop:step]

# df.loc[row_slicing, col_slicing]
# df.loc[start:stop:step, start:stop:step]

# (stop_index is included)
```

```python
df
```

```python
# First 3 rows

df.iloc[:3]
```

```python
df.loc['a':'c']
```

```python
# Reverse the rows

df.iloc[::-1]
```

```python
df.loc['h':'a':-1]
```

```python
df
```

```python
# First two columns
df.iloc[::, :2]
```

```python
# From ID all the columns for all rows

df.loc[::, 'ID':]
```

```python
### Reverse the columns

df.iloc[::, ::-1]
```
