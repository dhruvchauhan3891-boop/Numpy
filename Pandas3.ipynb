```python
import pandas as pd

df = pd.DataFrame({'Team' : ['A','A','B','B','C','C'],
                    'Players' : ['P1','P2','P3','P4','P5','P6'],
                    'Score' : [15,18,10,25,20,23]})
df
```

```python
df['Team_Avg_Score'] = df.groupby('Team')['Score'].transform('median')
df
```

```python
grouped = df.groupby('Team').sum()
grouped
```
