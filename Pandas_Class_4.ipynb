```python
dur='years'
def clean_duration(dur):
    mt=re.search(r'[0-9.]+', dur)
    if mt:
        return mt.group()
    else:
        return dur
```

### how='inner'

```python
# - Inner join
# - It takes common observations from both teh data frames for the common columns.

# emp- 10,20,30,50
# dt- 10,20,30,40

# inner- 10,20,30
```

```python
pd.merge(emp, dt, on='Depart_No', how='inner')
```

```python
pd.merge(emp, dt, on='Depart_No')
```

### how='left'

```python
# - It takes common observations from the left data frame for the common columns.

# emp- 10,20,30,50
# dt- 10,20,30,40

# left- 10,20,30,50
```

```python
pd.merge(emp, dt, on='Depart_No', how='left')
```

### how='right'

```python
# - It takes common observations from the right data frame for the common columns.

# emp- 10,20,30,50
# dt- 10,20,30,40

# left- 10,20,30,40
```

```python
pd.merge(emp, dt, on='Depart_No', how='right')
```

```python
dt
```

```python
emp
```

### how='outer'

```python
# - It takes all the observations from both data frame for the common columns.

# emp- 10,20,30,50
# dt- 10,20,30,40

# left- 10,20,30,40,50
```

```python
pd.merge(emp, dt, on='Depart_No', how='outer')
```

```python
pd.merge(emp, dt, on=['Depart_No', 'City'], how='inner')
```
